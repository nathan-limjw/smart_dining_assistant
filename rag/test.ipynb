```python
import re
def normalize_name(name):
    return re.sub(r"[^\w\s]", "", name.lower()).strip()

normalized_cities = {normalize_name(c): c for c in pa_cities}
normalized_cities_names = list(normalized_cities.keys())
```

```python
import re
import json
import difflib
import pandas as pd
# 2️⃣ Build similarity graph (ignore admin tokens when testing overlap)
ADMIN_TOKENS = {"township", "borough", "county", "pa", "twp", " valley"}


def jaccard(a, b):
    ta = set(a.split()) - ADMIN_TOKENS
    tb = set(b.split()) - ADMIN_TOKENS
    inter = len(ta & tb)
    union = len(ta | tb)
    return inter / union if union else 0

def are_related(a, b):
    sim = difflib.SequenceMatcher(None, a, b).ratio()
    jacc = jaccard(a, b)
    # must be either *very* similar or share >1 non-admin token
    return (sim > 0.88) or (jacc >= 0.5 and len(set(a.split()) & set(b.split())) > 1)


groups = []
visited = set()
for i, n1 in enumerate(normalized_cities_names):
    if n1 in visited:
        continue
    group = {n1}
    for n2 in normalized_cities_names:
        if n1 == n2 or n2 in visited:
            continue
        sim = difflib.SequenceMatcher(None, n1, n2).ratio()
        # require real (non-admin) token overlap
        if are_related(n1,n2):
            group.add(n2)
            visited.add(n2)
    groups.append(group)


groups 
```

```python
def strip_admin_suffix(name):
    name=name.lower()
    for suffix in [" township", " borough", " county", " pa", " twp", " valley"]:
        if name.endswith(suffix):
            name=name.replace(suffix, "")
    return name.strip()

aliases={}
for group in groups:
    originals = [normalized_cities[g] for g in group]
    cleaned = [strip_admin_suffix(o) for o in originals]

    # --- Rule 1: if there is a multi-word name that contains >1 unique single-word members, prefer that ---
    multi_word = [o for o in originals if len(o.split()) > 1]
    single_word = [o for o in originals if len(o.split()) == 1]

    if multi_word and len(single_word) > 1:
        # choose the multiword that contains all singles' tokens
        best_multi = None
        for m in multi_word:
            tokens = set(m.lower().split())
            if all(s.lower() in tokens for s in single_word):
                best_multi = m
                break
        canonical = best_multi if best_multi else max(multi_word, key=len)

    # --- Rule 2: if a parent shorter name is substring of a longer one (Philadelphia case) ---
    elif any(short.lower() in long.lower() and short != long for short in originals for long in originals):
        canonical = min(originals, key=len)

    # --- Rule 3: else if multi-word exists but no parent-child relation, pick longest multi-word ---
    elif multi_word:
        canonical = max(multi_word, key=len)

    # --- Rule 4: fallback to shortest base name ---
    else:
        canonical = min(originals, key=len)

    for g in group:
        aliases[g] = canonical

aliases
```

```python
import os
DATA_PATH=os.getcwd()
alias_path = os.path.join(DATA_PATH, "city_aliases.json")

with open(alias_path, "w") as f:
    json.dump(aliases, f, indent=2)

print(f"saved to {alias_path}")
```
